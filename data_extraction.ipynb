```python
import pandas as pd 
import numpy as np
```

# Loading and cleaning data

```python
# Loading the second sheet of the Excel file
data = pd.read_excel('data/data.xlsx', sheet_name=2)
data
```

```python
# Drop rows for which we have no company name
data = data[~data["Company Name"].isna()]
```

```python
# Dropping the first column which is the index
data.drop(columns=["Unnamed: 0"], inplace=True)
data.head()
```

```python
data.columns
```

```python
# No information in the following columns, columns used for previous filtering
print(data["Disponibilités données financières"].value_counts())
print(data["Perte moyenne sur 3 ans"].value_counts())
print(data["Selected companies"].value_counts())
print(data["Reason for rejection"].value_counts())
print(data["2 années de pertes consécutives"].value_counts())
```

```python
# Let us drop the columns that we won't be using   
columns_to_drop = [
    "Description in Original language",
    "Disponibilités données financières",
    "Perte moyenne sur 3 ans",
    "Selected companies",
    "Reason for rejection",
    "2 années de pertes consécutives"
]
data = data.drop(columns=columns_to_drop)
data.tail()
```

# Extracting the comparable companies

```python
## Manufacturing & design companies
manuf_and_design_comps = data[data["Main activity"].str.contains(r"Manufacturing|Design", case=False, na=False)]
manuf_and_design_comps
```

Maybe we could keep companies that also do retail

```python
# Let us drop these columns
data.drop(manuf_and_design_comps.index, axis=0, inplace=True)
print(f"Number of companies after dropping manufacturing and design companies: {data.shape[0]}")
data.head()
```

```python
# Exclude companies in sportswear
# DO NOT DELETE THE FOLLOWING CODE
# words_to_exclude = input("Words to exluce") # Here input "Sportwear, Sports, Sport"
# words_to_exclude_sep = [word.strip() for word in words_to_exclude.split(",")]
# words_to_exclude_final = []
# for word in words_to_exclude_sep:
#     if len(words_to_exclude_final)>0:
#         words_to_exclude_final += "|" + word
#     else:
#         words_to_exclude_final = word
# sportswear_companies = data[data["Description in English"].str.contains(r"words_to_exclude_final", case=False, na=False)]
sportswear_companies = data[data["Description in English"].str.contains(r"Sportswear|Sports|Sport", case=False, na=False)]
sportswear_companies
```

```python
# Excluse these rows from the data
data.drop(sportswear_companies.index, axis=0, inplace=True)
print(f"Number of companies after dropping sportswear companies: {data.shape[0]}")
data
```

```python
# Using an LLM on the description to classify the companies as 'LUXURY' or 'NOT LUXURY'
# Takes 1-2 minutes to run
import ollama

model_name = "llama3.2"

def classify_luxury(desc):
    response = ollama.chat(
        model=model_name,
        messages=[{"role": "user", "content": f"Classify this as 'LUXURY' or 'NOT LUXURY': {desc}. Just output LUXURY or NOT LUXURY"}]
    )
    return response['message']['content']

# Apply function to create the 'luxury' column
data["luxury"] = data["Description in English"].apply(classify_luxury)
```

```python
# Check the results
data["luxury"].value_counts()
```

```python
# Here are the companies classified as 'LUXURY'
luxury_companies = data[data["luxury"] == "LUXURY"]
luxury_companies
```
