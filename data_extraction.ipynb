# Second approach

In this approach, we use llama3.2 to extract the name of the brand from the website as the name of the company is not easily usable.

```python
data = pd.read_csv("data/data_cleaned.csv")
data.head()
```

First we use llama3.2 to extract the brand name so we can use that instead of the company name which is sometimes less informative.

```python
# We retrieve the brand name from the website of the company
# Takes 1-2 minutes to run
import ollama

model_name = "llama3.2"

def classify_luxury(desc):
    if desc == np.nan:
        return np.nan
    else:
        response = ollama.chat(
            model=model_name,
            messages=[{"role": "user", "content": f"Identify the brand on this website: {desc}. Just output the brand name, or what looks like a brand name, no need to write anything else. If you cannot identify the brand name, just write 'NaN'."}],
        )
        return response['message']['content']

# Apply function to create the 'luxury' column
data["identified_brand"] = data["Website"].apply(classify_luxury)
data[["Company Name", "Website", "identified_brand"]]
```

We further clean the identified_brand 

```python
# Additional cleaning of identified_brand
data["identified_brand_clean"] = data["identified_brand"].str.split("\n").str[0]
# replace name with Nan if it len(identified_brand_clean) >30 meaning it is a sentence
data.loc[data["identified_brand_clean"].str.len() > 30, "identified_brand_clean"] = np.nan
# Replace Nan with np.nan
data.loc[data["identified_brand_clean"] == "NaN", "identified_brand_clean"] = np.nan
data[["Company Name", "Website", "identified_brand_clean"]]
```

Now, we ask llama3.2 what it knows about Fendi.

```python
model_name = "llama3.2"
response = ollama.chat(
        model=model_name,
        messages=[{"role": "user", "content": "What is Fendi, the company?"}]
    )
print(response['message']['content'])
```

llama3.2 seems to know quite a lot about Fendi, so let us use this knowledge instead of giving it a less detailed definition of Fendi to identify comparables.

I suggest that you do not rerun the following cell as it takes about 10m to run.

```python
model_name = "llama3.2"

def classify_luxury(row):
    name = row["identified_brand_clean"]
    desc = row["Description in English"]
    response = ollama.chat(
        model=model_name,
        messages=[{"role": "user", "content": f"Using your knowledge of the company {name}, and this description ({desc}), classify it as working in the same industry as Fendi. Just output YES or NO. Then the reason in one sentence."}]
    )
    output = response['message']['content'].split(".", 1)
    luxury = output[0].strip()  # YES or NO
    reason = output[1].strip() if len(output) > 1 else ""  # The rest is the reason
    luxury_split = luxury.split("\n", 1)
    reason = luxury_split[1] if len(luxury_split) > 1 else reason
    # if name == "Zara":
    print(luxury, reason)

    return luxury, reason

data[["luxury", "reason"]] = pd.DataFrame(
    data.apply(classify_luxury, axis=1).to_list(),
    index=data.index
)
```

```python
# This approach helps us both to tackle this challenge efficiently and to have a clear understanding of the reasons behind the classification.
data.head()
```

```python
luxury_comps = data[data["luxury"]=="YES"]
print(f"{luxury_comps.shape[0]} companies have been classified as luxury companies")
```
