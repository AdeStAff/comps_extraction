# Second approach

In this approach, we use llama3.2 to extract the name of the brand from the website as the name of the company is not easily usable.

```python
data = pd.read_csv("data/data_cleaned.csv")
data.head()
```

```python
# Using an LLM on the description to classify the companies as 'LUXURY' or 'NOT LUXURY'
# Takes 1-2 minutes to run
import ollama

model_name = "llama3.2"

def classify_luxury(desc):
    if desc == np.nan:
        return np.nan
    else:
        response = ollama.chat(
            model=model_name,
            messages=[{"role": "user", "content": f"Identify the brand on this website: {desc}. Just output the brand name, or what looks like a brand name, no need to write anything else. If you cannot identify the brand name, just write 'NaN'."}],
        )
        return response['message']['content']

# Apply function to create the 'luxury' column
data["identified_brand"] = data["Website"].apply(classify_luxury)
data[["Company Name", "Website", "identified_brand"]]
```

```python
# Additional cleaning of identified_brand
data["identified_brand_clean"] = data["identified_brand"].str.split("\n").str[0]
# replace name with Nan if it len(identified_brand_clean) >30 meaning it is a sentence
data.loc[data["identified_brand_clean"].str.len() > 30, "identified_brand_clean"] = np.nan
# Replace Nan with np.nan
data.loc[data["identified_brand_clean"] == "NaN", "identified_brand_clean"] = np.nan
data[["Company Name", "Website", "identified_brand_clean"]]
```

```python
model_name = "llama3.2"
response = ollama.chat(
        model=model_name,
        messages=[{"role": "user", "content": "What is Fendi, the company?"}]
    )
print(response['message']['content'])
```

```python
model_name = "llama3.2"

def classify_luxury(desc):
    response = ollama.chat(
        model=model_name,
        messages=[{"role": "user", "content": f"Classify this company with the description: {desc} as a comparable of Fendi. Just output YES or NO. Then the reason in one sentence."}]
    )
    output = response['message']['content'].split(".", 1)
    luxury = output[0].strip()  # YES or NO
    reason = output[1].strip() if len(output) > 1 else ""  # The rest is the reason
    return luxury, reason

data[["luxury", "reason"]] = data["Description in English"].apply(classify_luxury).apply(pd.Series)
```

```python
data[["Company Name", "Website", "identified_brand_clean", "luxury", "reason"]]
```

```python
data["luxury_clean"] = data["luxury"].str.split("\n", n=1).str[0]

if data["reason"].isna().all():
    data["reason_clean"] = data["luxury"].str.split("\n", n=1).str[1]
else:
    data["reason_clean"] = data["reason"]

data[["Company Name", "Website", "identified_brand_clean", "luxury_clean", "reason_clean"]]
```
